```python
# Función para la plantilla del juego
from IPython.display import clear_output
import random
import sys

def display_board(board):
    
    print('{0:^4} | {1:^4} | {2:^4}'.format(board[0],board[1],board[2]))
    print('-----------------')
    print('{0:^4} | {1:^4} | {2:^4}'.format(board[3],board[4],board[5]))
    print('-----------------')
    print('{0:^4} | {1:^4} | {2:^4}'.format(board[6],board[7],board[8]))
    
def player_input():
    while True:
        player = input("Please pick a marker 'X' or 'O'")
        if player.upper() == 'X' or player.upper() == 'O':
            break
        
    return player.upper()

def place_marker(board, marker, position):
    board[position-1] = marker
    
def win_check(board, mark):
    test_string = ''.join(board)
    if mark*3 in test_string[::4] or mark*3 in test_string[2:7:2]:
        return True
    
    for e in range(0,7,3):
        if mark*3 in test_string[e:e+3]:
            return True
    for e in range(0,3):
        if mark*3 in test_string[e::3]:
            return True
    else:
        return False


import random

def choose_first():
    first_player = random.randint(1,10)
    second_player = random.randint(1,10)
    print(f'First player rolled... ¡{first_player}!\nSecond Player rolled... ¡{second_player}!')
    if first_player > second_player:
        return print('First player starts the game!')
    else:
        return print('Second player starts the game!')
    
def space_check(board, position): return board[position-1] == ' '

def full_board_check(board): return ' ' not in board

def player_choice(board):
    while True:
        try:
            pos = int(input('Ingrese un valor entre 1-9: '))
            if pos > 0 and pos < 10:
                if space_check(board, pos):
                    break
                else:
                    print('Espacio ocupado...')
        except ValueError:
            print('Insert a Number value')   
    return pos
    
def replay():
    play = input('do you wanna play again? (y/n)')
    if play.lower().startswith('y'):
        return True
    else:
        return False
```

```python
def main():
    print('Welcome to Tic Tac Toe!')
    p1Market=''
    p2Market=''
    game_on = True
    while True:
        boardGame = [' ',' ',' ',' ',' ',' ',' ',' ',' ']
        print('Lets see WHO PLAYER STARTS FIRST!')
        choose = choose_first()
        p1Market= player_input()
        if p1Market == 'X':
            p2Market = 'O'
        else:
            p2Market ='X'
        clear_output()      
        while game_on:
            display_board(boardGame)
            p1_value = player_choice(boardGame)
            place_marker(boardGame,p1Market,p1_value)
            clear_output()
            sys.stdout.flush()
            display_board(boardGame)
            if win_check(boardGame,p1Market):
                print('Player 1 has won the game')
                break

            if not full_board_check(boardGame):   
                p2_value = player_choice(boardGame)
                place_marker(boardGame,p2Market,p2_value)
                clear_output()
                if win_check(boardGame,p2Market):
                    display_board(boardGame)
                    print('Player 2 has won the game')
                    break
            else:
                print('¡EQUALS!')
                break


        if not replay():
            clear_output()
            print('Bye Bye...')
            break
        clear_output()
```

```python
main()
```
